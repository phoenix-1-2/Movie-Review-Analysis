```python
def simple(tag):
    if tag.startswith('J'):
        return 'a'
    if tag.startswith('R'):
        return 'r'
    if tag.startswith('A'):
        return wordnet.ADV
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('V'):
        return wordnet.VERB
    return wordnet.NOUN
```

```python
Y=[]
X=[]
for catagory in movie_reviews.categories():
    for ids in movie_reviews.fileids(catagory):
        Y.append(catagory)
        X.append( clean_doc(movie_reviews.words(ids)) )
```
